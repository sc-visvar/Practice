```python
import re
```

```python
usernames = [
    "Alice123",
    "Bob_456",
    "Charlie789 child",
    "123David zone",
    "Eve_321",
    "Frank007",
    "Grace42 temp",
    "Hank_99",
    "Ivy2020",
    "Jack_2001",
    "K1evin",
    "L0gan",
    "Miley5",
    "Nate_73",
    "Oscar6",
    "P4ul",
    "Quincy_88",
    "Rachel_101",
    "Steve_007",
    "Tommy202",
    "Uma2 zin",
    "Victor_93",
    "Wendy_7 Car",
    "Xavier_55",
    "Yara300",
    "Zane_47 Joke",
    "https://www.google.com"
]
```

```python
def search(regex):
    for username in usernames:
        result = re.search(regex, username)
        if result:
            print("Username: " + username)
            print("Matched substring: " + result.group())
            print("Span of the substring in the string: " + str(result.span()) + '\n')
```

```python
print("Searching for names which has first and last name..\n")

two_letter_regex = r'^\w+\s+\w+$' 
search(two_letter_regex)
```

```python
print("Seaching for word char sequence starting with C..\n")
sequence_with_C_regex = r'C\w+'
search(sequence_with_C_regex)
```

```python
print("Dividing and accessing the groups in strings...\n")
divide_access_groups_regex = r'^(?P<first>\w+)\s+(?P<second>\w+)$'
for username in usernames:
    result = re.search(divide_access_groups_regex, username)
    if result:
        print("Username: " + username)
        print("Matched substring's first group: " + result.group('first'))
        print("Matched substring's second group: " + result.group('second'))
        print("Span of the substring in the string: " + str(result.span()) + '\n')
```

```python
print("Searching for blocks of lower letters...\n")
small_regex = r'[a-z]+'
search(small_regex)

print("Findall in the strings..\n")
for username in usernames:
    result = re.findall(small_regex, username)
    if result:
        print(result)

# Use **re.finditer** to access to each object in the string with the property
```

```python
print("Match the strings, that is check the start of the string without using ^...\n")
match_regex = 'Ch?'
for username in usernames:
    if re.match(match_regex, username):
        print(username)

print("\nUsing fullmatch to seach for complete string..\n")
full_regex = r'https?://w{3}.\w+.(org|com)'
for username in usernames:
    if re.fullmatch(full_regex, username):
        print(username)
```
